Set-Up

```python
import os
import math
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
from itertools import product
from joblib import Parallel, delayed

# control the figure size
plt.rcParams["figure.figsize"] = (20,3)

# Load demand data
demand_data = pd.read_csv(
    'https://raw.githubusercontent.com/nosadchiy/public/main/demand.csv',
    dtype={"Month": str, "Date": int, "Demand": int}
)

# load intake price data
intake_price_data =  pd.read_csv(
    'https://raw.githubusercontent.com/nosadchiy/public/main/intake_price.csv',
    dtype={"Intake": int, "Price": float})

# decision variables
calng_price = 7.50 # set by CaLNG

peak_intake_from_pipeline = 181
```

Utility

```python
def determine_actual_usage(demand_data, calng_price, peak_intake_from_pipeline, total_lng_intake):
    usage_data = demand_data.copy()

    # Add price index column
    usage_data.reset_index(inplace=True, drop=False)
    usage_data = usage_data.rename(columns={'index': 'price_index'})

    # Compute pipeline price
    usage_data['pipeline_price'] = np.exp(np.maximum(0, (usage_data['Demand'] - 144) / 60))

    # Compute cost if using only pipeline
    usage_data['cost_from_pipeline_only'] = usage_data['Demand'] * usage_data['pipeline_price']

    # Compute demand for LNG
    usage_data['demand_for_lng'] = np.maximum(usage_data['Demand'] - peak_intake_from_pipeline, 0)

    # Cumulative LNG usage logic
    cumulative_lng_purchase = np.cumsum(usage_data['demand_for_lng']).shift(fill_value=0)
    usage_data['purchase_from_calng'] = np.minimum(usage_data['demand_for_lng'], total_lng_intake - cumulative_lng_purchase)

    # Compute pipeline intake and its price
    usage_data['pipeline_intake'] = usage_data['Demand'] - usage_data['purchase_from_calng']
    usage_data['pipeline_intake_price'] = np.exp(np.maximum(0, (usage_data['pipeline_intake'] - 144) / 60))

    # Compute total cost
    usage_data['cost_from_both'] = (
        usage_data['pipeline_intake'] * usage_data['pipeline_intake_price']
        + usage_data['purchase_from_calng'] * calng_price
    )

    # Compute savings
    usage_data['savings'] = usage_data['cost_from_pipeline_only'] - usage_data['cost_from_both']

    return usage_data

# Get usage data
usage_data = determine_actual_usage(demand_data, calng_price, peak_intake_from_pipeline, 1288)
```

```python
def calculate_cost_statistics(usage_data, calng_price, peak_intake_from_pipeline, total_lng_purchase, return_data=False):
    cost_data = usage_data.copy()
    cost_data['pipeline_intake_cost'] = cost_data['pipeline_intake'] * cost_data['pipeline_intake_price']
    cost_data['calng_intake_cost'] = cost_data['purchase_from_calng'] * calng_price

    total_pipeline_intake = cost_data['pipeline_intake'].sum()
    total_calng_intake = cost_data['purchase_from_calng'].sum()

    total_pipeline_intake_cost = cost_data['pipeline_intake_cost'].sum() * 1000
    total_calng_intake_cost = total_lng_purchase * calng_price * 1000
    total_utility_cost = total_pipeline_intake_cost + total_calng_intake_cost
    total_savings = cost_data['savings'].sum() * 1000

    average_demand = cost_data['Demand'].mean()
    average_price = np.exp(max(0, (average_demand - 144) / 60))

    total_average_draw_daily = cost_data['Demand'].sum() * average_price * 1000

    if not return_data:
        print(f"CaLNG price: ${calng_price:<04}")
        print("---")
        print(f"Peak intake from pipeline: {peak_intake_from_pipeline:,.0f} MMcf")
        print(f"Total peak months intake from pipeline: {total_pipeline_intake:,.0f} MMcf")
        print(f"Total intake from CaLNG: {total_calng_intake:,.0f} MMcf")
        print("---")
        print(f"Total amount paid to pipeline: ${total_pipeline_intake_cost:,.0f}")
        print(f"Total amount paid to CaLNG: ${total_calng_intake_cost:,.0f}")
        print(f"Total cost for utility: ${total_utility_cost:,.0f}")
        print(f"Savings from using CaLNG: ${total_savings:,.0f}")
        print("---")
        print(f"Total cost if average draw daily: ${total_average_draw_daily:,.0f}")
    else:
        return {
            "peak_intake_from_pipeline": peak_intake_from_pipeline,
            "total_lng_purchase": total_lng_purchase,
            "total_pipeline_intake": total_pipeline_intake,
            "total_calng_intake": total_calng_intake,
            "total_pipeline_intake_cost": total_pipeline_intake_cost,
            "total_calng_intake_cost": total_calng_intake_cost,
            "total_utility_cost": total_utility_cost,
            "total_savings": total_savings,
            "total_average_draw_daily": total_average_draw_daily
        }
```

Option Optimization

```python
def simulate_scenario(total_lng_intake, peak_intake_from_pipeline, calng_price, demand_data):
    usage_data = determine_actual_usage(demand_data, calng_price, peak_intake_from_pipeline, total_lng_intake)
    return calculate_cost_statistics(usage_data, calng_price, peak_intake_from_pipeline, total_lng_intake, return_data=True)

num_draws = 5000
total_lng_range = range(700, 1300, 50)
peak_intake_from_pipeline_range = range(170, 200, 5)

param_combinations = list(product(total_lng_range, peak_intake_from_pipeline_range))

# Parallel execution
results = Parallel(n_jobs=-1)(
    delayed(simulate_scenario)(total_lng, peak_intake, calng_price, demand_data.copy())
    for total_lng, peak_intake in param_combinations
)

# Convert results to DataFrame
results_df = pd.DataFrame(results)

# Check the columns of the resulting DataFrame
print(results_df.columns)

# Find the optimal scenario by minimizing total utility cost
optimal_scenario = results_df.loc[results_df['total_utility_cost'].idxmin()]
optimal_lng = optimal_scenario["total_lng_purchase"]
optimal_peak = optimal_scenario["peak_intake_from_pipeline"]
optimal_cost = optimal_scenario["total_utility_cost"]

print(f"Optimal total_LNG: {optimal_lng}")
print(f"Optimal peak_intake_from_pipeline: {optimal_peak}")
```

Plots

```python
# Compute cost stats for total_LNG variation (fixed peak intake)
df_lng = results_df[results_df["peak_intake_from_pipeline"] == optimal_peak]
mean_lng = df_lng.groupby("total_lng_purchase")["total_utility_cost"].mean()
std_lng = df_lng.groupby("total_lng_purchase")["total_utility_cost"].std()

# Compute cost stats for peak intake variation (fixed total LNG)
df_peak = results_df[results_df["total_lng_purchase"] == optimal_lng]
mean_peak = df_peak.groupby("peak_intake_from_pipeline")["total_utility_cost"].mean()
std_peak = df_peak.groupby("peak_intake_from_pipeline")["total_utility_cost"].std()

# ---- PLOTS ----
sns.set_style("whitegrid")

# Plot: Cost vs Total LNG
plt.figure(figsize=(10, 5))
plt.plot(mean_lng.index, mean_lng, label="Average Cost", color="b")
plt.fill_between(
    mean_lng.index,
    (mean_lng - std_lng).fillna(mean_lng),  # Avoid NaN issues
    (mean_lng + std_lng).fillna(mean_lng),
    alpha=0.2, color="b", label="±1 Std Dev"
)
plt.axvline(optimal_lng, color="k", linestyle="--", label="Optimal LNG Intake")
plt.xlabel("Total LNG Intake")
plt.ylabel("Total Cost")
plt.title("Cost vs Total LNG Intake (Fixed Peak Intake)")
plt.legend()
plt.show()

# Plot: Cost vs Peak Intake
plt.figure(figsize=(10, 5))
plt.plot(mean_peak.index, mean_peak, label="Average Cost", color="r")
plt.fill_between(
    mean_peak.index,
    (mean_peak - std_peak).fillna(mean_peak),  # Avoid NaN issues
    (mean_peak + std_peak).fillna(mean_peak),
    alpha=0.2, color="r", label="±1 Std Dev"
)
plt.axvline(optimal_peak, color="k", linestyle="--", label="Optimal Peak Intake")
plt.xlabel("Peak Intake from Pipeline")
plt.ylabel("Total Cost")
plt.title("Cost vs Peak Intake from Pipeline (Fixed Total LNG)")
plt.legend()
plt.show()
```



Given your analysis, does it make sense for a Utility company to sign a take-or-pay contract with CaLNG?

Yes, a take-or-pay contract could provide significant risk mitigation and cost savings. In particular, it lowers price volatility – pipeline gas may be very cheap on off‐peak days but can spike dramatically (above $10/Mcf) on peak demand days whereas locking in a $7.50/Mcf LNG option protects against these extreme spikes. The contract also ensures continuous supply when most needed, avoiding situations with steep unexpected charges (e.g., the Southern California Edison example), reducing the risk of sudden, high-cost purchases from pipelines being passed down to customers. Finally, peak-shaving simultaneously under multiple utilities drops overall pipeline intake, stabilizing and lowering market/spot price for everyone.

How would you structure the supply chain (i.e., use on-demand delivery or local storage) depending on the size of utility?

For large utilities with substantial peak demand, a local storage strategy, such as the satellite tank, can be especially advantageous. Although it requires a significant upfront capital investment for tanks and permits, building onsite storage enables these utilities to gradually accumulate inventory during off-peak seasons. As a result, they need fewer trailers available for just-in-time delivery during peak periods. Having local storage also provides immediate access to LNG when demand spikes, which is vital in avoiding the volatility and high prices that often arise when depending solely on pipeline gas during extreme weather. Despite higher initial costs, the overall long-term savings from requiring fewer trailers and mitigating exposure to exponential price hikes generally justify this approach for large utilities.

For smaller utilities with comparatively modest peaks, an on-demand delivery approach may be more sensible. Because these utilities do not need to manage massive daily swings in consumption, the cost of building and maintaining local storage tanks could exceed the benefits. They can rely on CaLNG’s trailer-based delivery to meet short-term spikes in demand, which avoids tying up capital in expensive infrastructure. While this approach offers less buffer on-site and may entail a slight risk if multiple customers need large LNG deliveries simultaneously, smaller utilities typically find the lower upfront costs outweigh the operational drawbacks.

What risks does CaLNG face?

CaLNG faces demand uncertainty, as the company’s business model depends heavily on utilities signing sizeable, long-term contracts. If fewer utilities commit or they opt for smaller contracts, CaLNG could be left with underutilized and expensive infrastructure. This is closely tied to price volatility risk: if pipeline infrastructure improves or other factors reduce price fluctuations, the guaranteed LNG price may seem less attractive to potential buyers. Moreover, regulatory hurdles—especially in a stringent environment like California—can delay or escalate the costs of constructing and operating LNG terminals and storage tanks.

CaLNG also faces operational and supply chain disruption risks, such as tanker delays, maintenance problems, or extreme weather events that might threaten its ability to uphold contractual obligations for guaranteed LNG delivery. Financing these large capital expenditures is another major challenge, given the possibility of rising interest rates or tighter credit conditions that could make such projects more expensive to fund. Finally, the entry of new competitors or the adoption of alternative technologies, such as large-scale battery energy storage, could reduce CaLNG’s market share over time.
